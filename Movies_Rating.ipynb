```python
#1 create a data frame and using is_null check if there is any null values in it
import pandas as pd
df=pd.read_csv("ratings.csv")
df
df.isnull().any()  # .any() is use to check if any null value in any column
# .sum() is use to check how many total of null value in each column
```

```python
#3 drop non values if present (there are no null values present in data set as we checked it before)
#4 drop duplicate records
df.duplicated().sum()
# there are no duplicated values in data frame ratings
```

```python
#5 as we change the data type of timestamp before
df=pd.read_csv("movies.csv")
df1=pd.read_csv("ratings.csv")
df1['timestamp'] = pd.to_datetime(df1['timestamp'],unit='s') # correcting the data type of timestamp from int64 to datetime64[ns]
print(df1)
```

```python
#6 create three new column name as year, month and quater
df1['year']=df1['timestamp'].dt.year
df1['month']=df1['timestamp'].dt.month
df1['quater']=df1['timestamp'].dt.quarter
df1
```

```python
#find out how many number of movies has been watched in each quater
mov=df1.groupby(['year','quater']).size()
mov
```

```python
# what are the average count of user watching movie per year
movyr=df1.groupby(['year','userId'])
movyr.size()
```

```python
# do a group by on movies column and find out the average rating and the total number of users who have watched that movie
# firstly we merge the movie and the ratings data frame for this question
import pandas as pd

merged_df = pd.merge(df, df1, on='movieId')
merged_df
```
